seabornのモジュールseaborn.matrixのヘルプ：

名前
    seaborn.matrix-データのマトリックスを視覚化する関数。

関数
    clustermap（data、pivot_kws = None、method = 'average'、metric = 'euclidean'、z_score = None、standard_scale = None、figsize = None、cbar_kws = None、row_cluster = True、col_cluster = True、row_linkage = None、col_linkage =なし、row_colors =なし、col_colors =なし、mask =なし、** kwargs）
        マトリックスデータセットを階層的にクラスター化されたヒートマップとしてプロットします。
        
        パラメーター
        ----------
        データ：2D配列のような
            クラスタリング用の長方形データ。 NAを含めることはできません。
        pivot_kws：dict、オプション
            `data`が整然としたデータフレームの場合、次のキーワード引数を提供できます
            ピボットして長方形のデータフレームを作成します。
        メソッド：str、オプション
            クラスターの計算に使用するリンケージ方法。
            詳細については、scipy.cluster.hierarchy.linkageのドキュメントを参照してください。
            https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html
        メトリック：str、オプション
            データに使用する距離メトリック。見る
            その他のオプションについては、scipy.spatial.distance.pdistのドキュメント
            https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.pdist.html
            行と列に異なるメトリック（またはメソッド）を使用するには、
            各リンケージ行列を自分で構築し、それらを提供します
            {row、col} _linkage。
        z_score：intまたはNone、オプション
            0（行）または1（列）のいずれか。 Zスコアを計算するかどうか
            行または列の場合。 Zスコアは次のとおりです。z=（x-平均）/ std
            各行（列）の値は、行（列）の平均を取得します
            減算し、行（列）の標準偏差で除算します。
            これにより、各行（列）の平均が0、分散が1になります。
        standard_scale：intまたはNone、オプション
            0（行）または1（列）のいずれか。それを標準化するかどうか
            各行または列の意味である次元は、最小値を引き、
            それぞれを最大値で割ります。
        figsize：2つの整数のタプル、オプション
            作成する図のサイズ。
        cbar_kws：dict、オプション
            `` heatmap``の `` cbar_kws``に渡すキーワード引数。に
            カラーバーにラベルを追加します。
        {row、col} _cluster：bool、オプション
            Trueの場合、{rows、columns}をクラスター化します。
        {row、col} _linkage：numpy.array、オプション
            行または列の事前計算されたリンケージマトリックス。見る
            特定の形式のscipy.cluster.hierarchy.linkage。
        {row、col} _colors：list-likeまたはpandas DataFrame / Series、オプション
            行または列のいずれかにラベルを付ける色のリスト。に有用
            グループ内のサンプルが一緒にクラスター化されているかどうかを評価します。できる
            複数の色レベルのラベル付けには、ネストされたリストまたはDataFrameを使用します。
            DataFrameまたはSeriesとして指定された場合、色のラベルが抽出されます
            DataFrames列名またはSeriesの名前から。
            DataFrame / Seriesの色も、データと一致します
            インデックス、色が正しい順序で描画されるようにします。
        mask：ブール配列またはDataFrame、オプション
            渡された場合、データは `` mask``がTrueのセルには表示されません。
            欠損値のあるセルは自動的にマスクされます。のみに使用
            計算ではなく視覚化。
        kwargs：他のキーワード引数
            他のすべてのキーワード引数は `` sns.heatmap``に渡されます
        
        戻り値
        -------
        clustergrid：ClusterGrid
            ClusterGridインスタンス。
        
        ノート
        -----
        返されたオブジェクトには `` savefig``メソッドがあります。
        樹形図をクリッピングせずにFigureオブジェクトを保存したい。
        
        並べ替えられた行インデックスにアクセスするには、次を使用します。
        `` clustergrid.dendrogram_row.reordered_ind``
        
        列インデックス、使用：
        `` clustergrid.dendrogram_col.reordered_ind``
        
        例
        --------
        
        クラスター化されたヒートマップをプロットします。
        
        ..プロット::
            ：context：close-figs
        
            >>> seabornをsnsとしてインポート。 sns.set（color_codes = True）
            >>> iris = sns.load_dataset（ "iris"）
            >>> species = iris.pop（ "species"）
            >>> g = sns.clustermap（iris）

            ：context：close-figs
        
            >>> g = sns.clustermap（iris、method = "single"）
        
        別のカラーマップを使用し、カラーマップの制限の外れ値を無視します。
        
        ..プロット::

モジュールseaborn.rcmodの関数plotting_contextのヘルプ：

plotting_context（context = None、font_scale = 1、rc = None）
    Figureの要素をスケーリングするためのパラメーターdictを返します。
    
    これは、ラベル、線、その他のサイズなどに影響します
    全体的なスタイルではなく、プロットの要素。基本コンテキスト
    「ノートブック」であり、他のコンテキストは「紙」、「トーク」、および「ポスター」です。
    これは、0.8、1.3、および1.6でスケーリングされたノートブックパラメーターのバージョンです。
    それぞれ。
    
    この関数は `` with``ステートメントで使用できるオブジェクトを返します
    コンテキストパラメータを一時的に変更します。
    
    パラメーター
    ----------
    コンテキスト：dict、None、または{paper、notebook、talk、poster}のいずれか
        パラメータの辞書または事前設定されたセットの名前。
    font_scale：float、オプション
        サイズを個別にスケーリングするための個別のスケーリング係数
        フォント要素。
    rc：dict、オプション
        プリセットシーボーンの値をオーバーライドするパラメーターマッピング
        コンテキスト辞書。これは、
        コンテキスト定義の一部と見なされます。
    
    例
    --------
    >>> c = plotting_context（ "poster"）
    
    >>> c = plotting_context（ "notebook"、font_scale = 1.5）
    
    >>> c = plotting_context（ "talk"、rc = {"lines.linewidth"：2}）
    
    >>> plpとしてmatplotlib.pyplotをインポート
